```python
# Importing Required Modules
import pandas as pd
import openpyxl
import numpy as np
```

```python
# READING EXCEL SHEET USING PANDAS
df = pd.read_excel("zenarate_user.xlsx", header=None)
df.shape
```

```python
# READING EXCEL SHEET USING OPENPYXL
from openpyxl import load_workbook
book = load_workbook('zenarate_user.xlsx')
```

```python
# GETTING ACTIVE SHEET
sheet = book.active
```

```python
# Creating Empty DataFrame for Storing Color Codes
df_color_codes = pd.DataFrame()

# get the cell color using the fill attribute of the cell
for i, row in enumerate(sheet.iter_rows()):
    for j, cell in enumerate(row):
        color = cell.fill.start_color.index
        df_color_codes.loc[i, j] = color
```

```python
# Saving ColorCoded DataFrame to CSV
df_color_codes.to_csv('zenarate_user_color_coded.csv', index=False, header=False)
```

```python
# Reading ColorCoded CSV
col_names = ['User Name', 'Role', 'LABEL\n ROLE', 'LABEL\n FPA Class', 'LABEL\n Network Office', 'Primary Coach',
 "Primary Coach's Email", 'Secondary Coach', "Secondary Coach's Email",'Other Coaches', "Other Coaches's Email",'Status','Date of Joining',
 'Date of Inactivation', 'User-Id', 'Secondary Email']

df2 = pd.read_csv('zenarate_user_color_coded.csv', names=col_names)
df2.shape
```

```python
df2.head()
```

```python
# Converting DataFrame To List of List
data_list = []
for index in range(len(df2)):
    row = df2.loc[index].values.flatten().tolist()
    data_list.append(row)
print(len(data_list), len(row))
```

```python
# Finding Rows that Have '00000000' in all columns and storing it in List.

perfect_rows = []
total_columns = 16
blank_color_code = '00000000'

for index in range(len(data_list)):
    row = data_list[index]
    count = row.count(blank_color_code)
    if int(count) == total_columns:
        perfect_rows.append(index)

perfect_rows
        
```

```python
# Dropping Perfect Rows From Orignal dataFrame, i.e. df here

df.drop(df.index[perfect_rows], axis=0, inplace=True)
df.shape
```

```python
# Saving Updated Dataframe In CSV format
df.to_csv('zenarate_user_cleaned.csv', index=False, header=False)
```
